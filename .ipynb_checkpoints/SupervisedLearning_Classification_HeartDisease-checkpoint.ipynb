# Supervised Learning - Classification
## Heart disease

Supervised classification - Jupyter Notebook of a classification problem, using the UCI heart disease data. 

## 1. Problem definition
Predict the probability of developing a heart disease giving the clinical parameter of a patient

## 2. Data 
Data taken from the UCI database https://archive.ics.uci.edu/ml/datasets/Heart+Disease
The data are present on Kaggle as well: https://www.kaggle.com/ronitf/heart-disease-uci

## 3. Evaluation

Accuracy will be used to evaluate the model

## 4. Features

Information on the features can be found here: https://www.kaggle.com/ronitf/heart-disease-uci and https://archive.ics.uci.edu/ml/datasets/Heart+Disease

* **age:** age in years
* **sex:** 
    * 1 = male
    * 0 = female
* **cp:** chest pain type
    * 1 = typical angina
    * 2 = atypical angina
    * 3 = non-anginal pain
    * 4 = asymptomatic
* **trestbps:** resting blood pressure (in mm Hg on admission to the hospital)
* **chol:** serum cholestoral in mg/dl
* **fbs:** fasting blood sugar > 120 mg/dl 
    * 1 = true
    * 0 = false
* **restecg:** resting electrocardiographic results
    * 0 = normal
    * 1 = having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
    * 2 = showing probable or definite left ventricular hypertrophy by Estes' criteria
* **thalach:** maximum heart rate achieved
* **exang:** exercise induced angina 
    * 1 = yes
    * 0 = no
* **oldpeak:** ST depression induced by exercise relative to rest
* **slope:** the slope of the peak exercise ST segment
    * 1 = upsloping
    * 2 = flat
    * 3 = downsloping
* **ca:** number of major vessels (0-3) colored by flourosopy
* **thal:** 
    * 3 = normal
    * 6 = fixed defect
    * 7 = reversable defect
* **target:** 
    * 1 = yes
    * 0 = no

### 5.3 Model selection

The following models will be tested to identify the best one: 
1. Logistic Regression
2. K-Nearest Neighbours Classifier
3. Random Forest Classifier

```python
# Create a dictionary with the different models
models = {"Logistic Regression": LogisticRegression(max_iter=1000), 
          "KNN": KNeighborsClassifier(), 
          "Random Forest": RandomForestClassifier()}
```

```python
# Create a function to fit and score the models
def fit_and_score(models, X_train, X_test, y_train, y_test):
    """
    Fits and evaluates given machine learning models
    models: dictionary of machine learning models
    X_train: training data (no labels)
    X_test: test data (no labels)
    y_train: training labels
    y_test: test labels    
    """
    # Create empty dictionary to record model scores
    model_scores = {}
    # Loop through models
    for name, model in models.items():
        # Fit the model
        model.fit(X_train, y_train)
        # Evaluate the model and append its score to the model_score dictionary
        model_scores[name] = round(model.score(X_test, y_test), 2)
    return model_scores    
```

```python
# Compare the models
model_scores = fit_and_score(models = models, 
                             X_train = X_train, 
                             X_test = X_test, 
                             y_train = y_train, 
                             y_test = y_test)
model_scores
```

```python
model_compare = pd.DataFrame(model_scores, index=['accuracy'])
model_compare
```

```python
T_model_compare = model_compare.T
T_model_compare
```

```python
sns.barplot(x=T_model_compare.index, y=T_model_compare['accuracy']);
```

*Logistic Regression* and *Random Forest Classifier* performs better, they will be brought forward in the next experimentation fase. 

## 6. Modelling

### 6.1 Hyperparameter tuning
Using RandomizedSearchCV to tune the parameters of the models

```python
# Import RandomizedSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Create a hyperparameter grid for LogisticRegression
log_reg_grid = {"C": np.logspace(-4, 4, 20), 
                "solver": ["liblinear"], 
                "penalty": ['l1', 'l2']}

# Create a hyperparameter grid for RandomForestClassifier
rfc_grid = {"n_estimators": np.arange(10, 1000, 50), 
           "max_depth": [None, 3, 5, 10], 
           "min_samples_split": np.arange(2, 20, 2), 
           "min_samples_leaf": np.arange(1, 20, 2)}
```

*Tune LogisticRegression model*

```python
# Set up random hyperparameter search for LogisticRegression
rs_log_reg = RandomizedSearchCV(LogisticRegression(), 
                                param_distributions=log_reg_grid, 
                                cv=5, 
                                n_iter=40)

# Fit random hyperparameter search model for LogisticRegression
rs_log_reg.fit(X_train, y_train)
```

```python
# Check the best parameters
rs_log_reg.best_params_
```

```python
# Evaluate on test data (best parameters are used by default)
round(rs_log_reg.score(X_test, y_test), 2)
```

No improvement compared to the previous model, without hyperparameters tuning. More parameters can be selected and tested. 

*Tune RandomForestClassifier*

```python
# Set up random hyperparameter search for RandomForestClassifier
rs_rfc = RandomizedSearchCV(RandomForestClassifier(), 
                            param_distributions=rfc_grid, 
                            cv=5, 
                            n_iter=50)

# Fit random hyperparameter search model for RandomForestClassifier
rs_rfc.fit(X_train, y_train)
```

```python
# Check the best parameters
rs_rfc.best_params_
```

```python
# Evaluate the model (with the best parameters) on test data
round(rs_rfc.score(X_test, y_test), 2)
```

Again, no improvement has been observed compared to the initial model tested. 

Since the LogisticRegression model performs better between the two, further hyperparameter optimisation will be performed using **GridSearchCV**

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Set up hyperparameter search for LogisticRegression using the previous grid
gs_log_reg = GridSearchCV(LogisticRegression(), 
                          param_grid=log_reg_grid, 
                          cv=5)

# Fit the grid hyperparameter search model
gs_log_reg.fit(X_train, y_train)
```

```python
# Check the best hyperparameters
gs_log_reg.best_params_
```

```python
# Evaluate with best params on test data
round(gs_log_reg.score(X_test, y_test), 2)
```

No score improvement, beginning model performed very well.

##### 6.2.1.1 ROC Curve and AUC score

The **ROC curve** is a graphical plot that illustrates the diagnostic ability of a binary classifier system as its discrimination threshold is varied.

The ROC curve is created by plotting the *true positive rate (TPR)* against the *false positive rate (FPR)* at various threshold settings. 

The **area under the curve (AUC)** is a measure of how well a parameter can distinguish between two diagnostic groups.


```python
from sklearn.metrics import plot_roc_curve
```

```python
# Plot ROC curve and calculate AUC metric
plot_roc_curve(gs_log_reg, X_test, y_test);
```

The AUC is 0.95. 

##### 6.2.1.2 Confusion matrix

The **Confusion matrix** is a performance measurement for machine learning classification problem where output can be two or more classes. It is a table with 4 different combinations of predicted and actual values.

More information here: https://towardsdatascience.com/understanding-confusion-matrix-a9ad42dcfd62

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_preds)
```

```python
# Visualise confusion matrix using seaborn 
sns.set(font_scale=1.5)

def plot_conf_mat(y_test, y_preds):
    """
    Plot confusion matrix using Seaborn heatmap
    """
    plt.figure(figsize=(3, 3))
    sns.heatmap(confusion_matrix(y_test, y_preds), 
                annot=True,
                cbar=False)
    plt.xlabel("True label")
    plt.ylabel("Predicted label")
    
plot_conf_mat(y_test, y_preds)
```

##### 6.2.1.3 Classification report

* **Precision:** percentage of correct predictions 
    * TP/(TP/FP)
* **Recall:** fractions of positives that were correctly identified 
    * TP(TP+FN)
* **F1 score:** percentage of correct positive predictions 
    * 2*(recall * precision)/(recall+precision)
* **Support:** number of occurences of the given class in the dataset (check for class imbalance)



* **Weighted average** is calculated taking into account the number of occurrences in each class

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_preds))
```

#### 6.2.2 Cross validation

Calculate precision, recall and f1-score of the model using cross validation and `cross_val_score()`

```python
from sklearn.model_selection import cross_val_score
```

```python
# Check the best hyperparameters
gs_log_reg.best_params_
```

```python
# Create a new classifier with the best parameters
clf = LogisticRegression(C=0.08858667904100823, 
                         penalty='l2', 
                         solver='liblinear')
```

```python
# Cross validated precision
cv_precision = cross_val_score(clf, 
                         X, 
                         y, 
                         cv=5, 
                         scoring='precision')
cv_precision=np.mean(cv_precision)
cv_precision
```

```python
# Cross validated recall
cv_recall = cross_val_score(clf, 
                         X, 
                         y, 
                         cv=5, 
                         scoring='recall')
cv_recall=np.mean(cv_recall)
cv_recall
```

```python
# Cross validated f1-score
cv_f1 = cross_val_score(clf, 
                         X, 
                         y, 
                         cv=5, 
                         scoring='f1')
cv_f1=np.mean(cv_f1)
cv_f1
```

```python
# Visualise cross validated metrics
cv_metrics = pd.DataFrame({"Accuracy": cv_acc, 
                           "Precision": cv_precision, 
                           "Recall": cv_recall, 
                           "f1": cv_f1}, index=['metrics'])
T_cv_metrics=cv_metrics.T
T_cv_metrics
```
